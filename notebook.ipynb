In this work, I will be trying to find interesting patterns in given data by University of Jordan. I have been supplied a dataset with an educational data set which was collected from learning management system. Data contains few mistakes which must be fixed before analysis.

The university wants to know and see visualizations for the following:
1. Students allocation by class.
2. Students allocation by country.
3. Students performance (by Class).
4. Who performs better, boys or girls?
5. Dependence in students performance and attendance.
6. Dependence in attendance and parentâ€™s satisfaction.
7. How students activity influences performance.

```python
import pandas as pd

df = pd.read_csv('5_workshop_data_1.csv')
df.head()
```

```python
df.columns = ['Gender','Nationality','PlaceofBirth','StageID','GradeID','SectionID','Topic','Semester','Relation','RaisedHands','VisitedResources','AnnouncementsView','Discussion','ParentAnsweringSurvey','ParentSchoolSatisfaction','StudentAbsenceDays','Class']
```

```python
df.head()
```

```python
df['Nationality'].unique()
```

```python
df['Nationality'] = df['Nationality'].replace('KW','Kuwait')
```

```python
df.head()
```

```python
df['Nationality'] = df['Nationality'].replace('lebanon','Lebanon')
df['Nationality'] = df['Nationality'].replace('venezuela','Venezuela')
df['Nationality'] = df['Nationality'].replace('Lybia','Libya')
df['Nationality'] = df['Nationality'].replace('SaudiArabia','Saudi Arabia')
```

```python
df.head()
```

```python
set(df['Class'])
```

```python
df['Class'] = df['Class'].replace('L','Low-Level')
df['Class'] = df['Class'].replace('M','Middle-Level')
df['Class'] = df['Class'].replace('H','High-Level')
```

```python
df.head()
```

```python
set(df['PlaceofBirth'])
```

```python
df['PlaceofBirth'] = df['PlaceofBirth'].replace('KuwaIT','Kuwait')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('lebanon','Lebanon')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('venezuela','Venezuela')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('Lybia','Libya')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('SaudiArabia','Saudi Arabia')
```

```python
set(df['StageID'])
```

```python
df['StageID'] = df['StageID'].replace('lowerlevel','LowerLevel')
```

```python
df.head()
```

---

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df.info()
```

### Students allocation by class

```python
sns.set_style("darkgrid")
plt.figure(figsize=(16,10))
sns.countplot(data = df, x = 'Topic')
plt.show()
```

### Students allocation by country

```python
plt.figure(figsize=(16,10))
sns.countplot(data = df, x = 'Nationality', order = df['Nationality'].value_counts().index)
plt.show()
```

### Students performance

```python
plt.figure(figsize=(16,10))
sns.countplot(data=df,x='Class',order=['Low-Level','Middle-Level','High-Level'],palette='Blues')
plt.show()
```

### Boys vs girls performance comparison

```python
plt.figure(figsize=(16,10))
sns.countplot(data=df,x='Gender',hue_order=['Low-Level','Middle-Level','High-Level'],hue='Class',palette='Blues')
plt.show()
```

### Dependence in students performance and attendance

```python
plt.figure(figsize = (16,10))
sns.countplot(data=df,x='StudentAbsenceDays',hue='Class',hue_order=['Low-Level','Middle-Level','High-Level'],palette='Blues')
plt.show()
```

### Dependence in attendance and parentâ€™s satisfaction

```python
plt.figure(figsize=(16,10))
sns.countplot(data=df,x='StudentAbsenceDays',hue='ParentSchoolSatisfaction',palette='Set2')
plt.show()
```

### How students activity influences performance

```python
plt.figure(figsize=(16,10))
sns.barplot(data=df,x='Class',y='RaisedHands',palette='Set1',order=['Low-Level','Middle-Level','High-Level'])
plt.show()
```
